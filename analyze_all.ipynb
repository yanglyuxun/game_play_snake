# The plots of the scores


```python
import matplotlib.pyplot as plt
import pickle
import numpy as np

def moving_avg(x,p=10):
    y = []
    for i in range(len(x)-p+1):
        y.append(np.mean(x[i:i+p]))
    return y
```

```python
with open('5_5/mem.pickle','rb') as f:
    (t5,_,_,scores5)=pickle.load(f)
with open('7_7/mem.pickle','rb') as f:
    (t7,_,_,scores7)=pickle.load(f)
with open('10_10/mem.pickle','rb') as f:
    (t10,_,_,scores10)=pickle.load(f)
```

```python
print('5*5: steps=%i, games=%i, max(score)=%i'%(t5,len(scores5),np.max(scores5)))
print('7*7: steps=%i, games=%i, max(score)=%i'%(t7,len(scores7),np.max(scores7)))
print('10*10: steps=%i, games=%i, max(score)=%i'%(t10,len(scores10),np.max(scores10)))
```

```python
plt.plot(scores5); plt.title('5*5'); plt.show()
plt.plot(scores7); plt.title('7*7'); plt.show()
plt.plot(scores10); plt.title('10*10'); plt.show()
```

```python
MA = 10
plt.plot(moving_avg(scores5,MA)); plt.title('5*5, MA(%i)'%MA); plt.xlabel('games'); plt.ylabel('score'); plt.show()
plt.plot(moving_avg(scores7,MA)); plt.title('7*7, MA(%i)'%MA); plt.xlabel('games'); plt.ylabel('score'); plt.show()
plt.plot(moving_avg(scores10,MA)); plt.title('10*10, MA(%i)'%MA); plt.xlabel('games'); plt.ylabel('score'); plt.show()
```

```python
MA = 50
plt.plot(moving_avg(scores5,MA)); plt.title('5*5, MA(%i)'%MA); plt.xlabel('games'); plt.ylabel('score')
plt.savefig('5_5/plot%i.png'%MA); plt.show()
plt.plot(moving_avg(scores7,MA)); plt.title('7*7, MA(%i)'%MA); plt.xlabel('games'); plt.ylabel('score')
plt.savefig('7_7/plot%i.png'%MA); plt.show()
plt.plot(moving_avg(scores10,MA)); plt.title('10*10, MA(%i)'%MA); plt.xlabel('games'); plt.ylabel('score')
plt.savefig('10_10/plot%i.png'%MA); plt.show()
```

```python
MA = 50
plt.plot(moving_avg(scores5,MA),label='5*5'); 
plt.plot(moving_avg(scores7,MA),label='7*7'); 
plt.plot(moving_avg(scores10,MA),label='10*10');
plt.xlim((0,min(len(scores5),len(scores7),len(scores10))))
plt.title('Learning processes, MA(%i)'%MA)
plt.legend()
plt.xlabel('games'); plt.ylabel('score')
plt.savefig('plot%i.png'%MA); plt.show()
```
