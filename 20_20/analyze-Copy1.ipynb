# The plots of the scores

changed epsilon to the initial value, and gave it another 100000 rather than 5000 steps to decrease, at the 6173th game.

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
with open('mem.pickle','rb') as f:
    (_,_,_,scores)=pickle.load(f)
```

```python
def moving_avg(x,p=10):
    y = []
    for i in range(len(x)-p+1):
        y.append(np.mean(x[i:i+p]))
    return y
```

```python
np.max(scores)
```

```python
plt.plot(scores)
```

```python
plt.plot(moving_avg(scores,3))
```

```python
plt.plot(moving_avg(scores,5))
```

```python
plt.plot(moving_avg(scores,10))
plt.xlabel('score')
plt.ylabel('games')
plt.savefig('plot.png')
```
